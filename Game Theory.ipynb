```python
import matplotlib.pyplot as plt
import numpy as np

# Define time points t0, t1, and t2
t0 = 0
t1 = 5
t2 = 10

# Generate time values from t0 to t2
time_values = np.linspace(t0, t2, 100)

# Initialize Age of Information (AoI) metrics
aoi_I = np.maximum(time_values - t0, 0)
aoi_II = np.zeros_like(time_values)

# At t1, AoII starts increasing due to a drift
aoi_II[time_values >= t1] = time_values[time_values >= t1] - t1

# At t2, status is refreshed
refreshed_time = t2
aoi_I[time_values >= refreshed_time] = time_values[time_values >= refreshed_time] - t0
aoi_II[time_values >= refreshed_time] = time_values[time_values >= refreshed_time] - t1

# Plot the Age of Information metrics
plt.plot(time_values, aoi_I, label='Data without Tampering', linewidth=2)
plt.plot(time_values, aoi_II, label='Data with Tampering', linewidth=2)

# Find nearest indices for t1 and t2
idx_t1 = np.argmin(np.abs(time_values - t1))
idx_t2 = np.argmin(np.abs(time_values - t2))

# Highlight key points
plt.scatter([t0, time_values[idx_t1], time_values[idx_t2]],
            [0, aoi_II[idx_t1], aoi_I[idx_t2]], color='red', zorder=5)
plt.text(t0, 0, f't0', ha='right', va='bottom')
plt.text(time_values[idx_t1], aoi_II[idx_t1], f't1', ha='right', va='top')
plt.text(time_values[idx_t2], aoi_I[idx_t2], f't2', ha='right', va='bottom')

# Add labels and legend

plt.title('Tampering')
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
from scipy.optimize import minimize

# Define the cost, alpha, and the expression to minimize
alpha = 0.1
cost = 1.0

# Define the objective function to minimize
def objective_function(p):
    return -(1 / (2 * (alpha * p) ** 2) / (1 / (alpha * p) + (1 / ((1 - alpha) * p)))) - cost * p

# Initial guess for the optimization algorithm
initial_guess = 1.0

# Minimize the objective function
result = minimize(objective_function, initial_guess)

# Extract the optimal value of p
optimal_p = result.x[0]

print(f"Optimal p: {optimal_p}")
print(f"Minimum value of the expression: {result.fun}")
```
